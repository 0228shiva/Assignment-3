```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned data
df = pd.read_csv(r'C:\Users\Sys\OneDrive\Desktop\New folder\rpa_inventory_project\data\clean_inventory.csv')

# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'])
df.info()
# 1. Sales Trend Over Time
plt.figure(figsize=(12, 6))
sales_by_date = df.groupby('date')['units sold'].sum()
sales_by_date.plot()
plt.title("Daily Units Sold Over Time")
plt.xlabel("Date")
plt.ylabel("Units Sold")
plt.grid(True)
plt.tight_layout()
plt.show()

# 2. Average Inventory Level by Region
plt.figure(figsize=(8, 5))
sns.barplot(x='region', y='inventory level', data=df, estimator='mean', ci=None)
plt.title("Average Inventory Level by Region")
plt.ylabel("Avg Inventory Level")
plt.tight_layout()
plt.show()

# 3. Units Sold by Product Category
plt.figure(figsize=(8, 5))
sns.boxplot(x='category', y='units sold', data=df)
plt.title("Units Sold by Product Category")
plt.tight_layout()
plt.show()

# 4. Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Between Numerical Features")
plt.tight_layout()
plt.show()

# 5. Weather Condition vs. Units Sold
plt.figure(figsize=(8, 5))
sns.barplot(x='weather condition', y='units sold', data=df, estimator='mean', ci=None)
plt.title("Avg Units Sold by Weather Condition")
plt.tight_layout()
plt.show()

# 6. Seasonality Distribution
plt.figure(figsize=(6, 4))
season_counts = df['seasonality'].value_counts()
season_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Seasonality Distribution")
plt.ylabel("")  # Hide y-axis label
plt.tight_layout()
plt.show()
```
